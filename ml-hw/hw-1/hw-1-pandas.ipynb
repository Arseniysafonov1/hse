### Визуализации и matplotlib

- у графика должно быть название;

#### 7. [0.5 балла] Постройте график (barplot) медианных перцентилей в зависимости от первого символа hex-кода id студента

Таким образом мы на глаз прикидываем, есть ли зависимость между, якобы, случайным id и значимым параметром. Если зависимость есть, и в дальнейшем мы будем сплитовать данные на тест и трейн по такому айдишнику без перемешивания, то можем допустить утечку, поэтому такие особенности важно отслеживать.

В графике явно задайте порядок столбцов по символам hex-алфавита (от 0 до f слева направо) и настройте отображение доверительных интервалов уровня **0.96**. Можно ли говорить о наличии проверяемой зависимости на основе визуальной информации?

```python
bar_data = df.copy()
bar_data['first_num'] = bar_data['id'].apply(lambda x : x[0])
bar_data['counter'] = 1
bar_data['percentile_std'] = bar_data['percentile']

bar_data = bar_data.sort_values('first_num').groupby('first_num')[['percentile', 'counter', 'percentile_std']] \
    .agg({'percentile' : 'median', 'counter' : 'sum', 'percentile_std' : 'std'})
bar_data
```

Сохраните график в формате pdf (так он останется векторизованным).

```python
# your code
import scipy.stats as ss

fig, ax = plt.subplots()

# https://stackoverflow.com/questions/15033511/compute-a-confidence-interval-from-sample-data
error = ss.t.ppf(0.96, bar_data['counter'] - 1) * bar_data['percentile_std']

ax.bar(bar_data.index, bar_data['percentile'], yerr=error)

ax.set_title("Медианные перцентили в зависимости от первого символа hex-кода id студента")

ax.set_xlabel('Первый символ id')
ax.set_ylabel('Перцентиль')

plt.savefig('barplot.pdf')

plt.show()
```

#### Перцентиль

В столбце `'percentile'` записан перцентиль студента в рамках своего курса своей ОП.

Перцентиль студента X определяется следующим образом: "процент студентов, чьи результаты обучения лучше, чем у студента X". Соответственно, 0% -- лучший студент на потоке, $\to$ 100% -- худший студент на потоке, то есть метрика убывает по "качеству" студента.

```python
def count_places(course):
    if course in ('Количественные финансы', 'Промышленное программирование на Haskell', 'Рекомендательные системы'):
        return 60
    if course == 'Глубинное обучение в обработке звука':
        return 1000
    if course == '':
        return 0
    return 30

fall_courses = pd.DataFrame()
fall_courses['course'] = fall[['c1', 'c2', 'c3']].values.tolist()
fall_courses = fall_courses.explode('course').groupby('course').first().reset_index()
fall_courses['left'] = fall_courses['course'].apply(count_places)
fall_courses = fall_courses.set_index('course')
```

```python
one_course = fall[fall['fall_course_num'] == 1]
two_course = fall[fall['fall_course_num'] == 2]

a1, a2, a3 = pd.DataFrame(), pd.DataFrame(), pd.DataFrame()

a1[['id', 'c1', 'c2', 'c3', 'percentile']] = one_course[['id', 'c1', 'c2', 'c3', 'percentile']]
a2[['id', 'c1', 'percentile']] = two_course[['id', 'c1', 'percentile']]
a3[['id', 'c1', 'percentile']] = two_course[['id', 'c2', 'percentile']]

a2['c2'] = two_course['c3']
a3['c2'] = two_course['c3']

a2['c3'] = ''
a3['c3'] = ''

attempts = pd.concat((a1, a2, a3))

ans = []
```

```python
for wave in range(3):
    attempts['course'] = attempts['c1']

    attempts_with_left = attempts.set_index('course').join(fall_courses).reset_index().sort_values('percentile')

    attempts_with_left['curr_rank'] = attempts_with_left.groupby('course').cumcount()
    attempts_with_left['is_success'] = attempts_with_left['curr_rank'] < attempts_with_left['left']
    
    ans.append(attempts_with_left[attempts_with_left['is_success']][['id', 'course']])

    if wave == 2:
        break

    left = attempts_with_left[attempts_with_left['is_success']][['course', 'left', 'curr_rank']]
    left['curr_rank'] += 1
    
    left = left.groupby('course').max()
    
    left = pd.DataFrame({'new_val': left['left'] - left['curr_rank']})
    
    left = fall_courses.join(left)
    left['left'] = left['new_val'].combine_first(left['left'])
    fall_courses = left[['left']]

    attempts = attempts_with_left[~attempts_with_left['is_success']][['id', 'c1', 'c2', 'c3', 'percentile']]
    attempts['c1'] = attempts['c2']
    attempts['c2'] = attempts['c3']
    attempts['c3'] = ''
    attempts.drop_duplicates(inplace=True)

ans = pd.concat(ans)
```
